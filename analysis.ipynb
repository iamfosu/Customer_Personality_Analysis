# Import and load data

```python
import pandas as pd
```

```python
df = pd.read_csv('Customer marketing.csv', sep='\t')
df.head()
```

```python
df.shape
```

```python
df.info()
```

# Rename columns to lower case

```python
#rename columns to lower case and uniform format
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

# Missing Value handling

```python
#Missing values
df.isnull().sum()
```

```python
# Handle missing values (Only income column has missing values)
df['income'] = df['income'].fillna(df['income'].median())
df['income'].isnull().sum()
```

```python
df.isnull().sum()
```

# Remove Duplicate rows

```python
# check duplicate values and remove duplicate rows
print("Duplicate values: ", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
print("Duplicate values after removal: ", df.duplicated().sum())
```

# Standardized the categorical data columns

```python
df['education'].value_counts()
```

```python
df['education'].unique()
```

Standardize the education column values

```python
# standardize the education column values
education_mapping = {
    'Graduation': 'Graduate', 
    'PhD': 'Postgraduate', 
    'Master': 'Postgraduate', 
    'Basic': 'Undergraduate', 
    '2n Cycle': 'Undergraduate'
}
df['education'] = df['education'].map(education_mapping)
```

```python
print(df['education'].unique())
print(df['education'].value_counts())
```

Standardize the Marital status column values

```python
print(df['marital_status'].unique())
print(df['marital_status'].value_counts())
```

```python
# standardize the Marital_status column values
marital_status_mapping = {
    'Single': 'Single', 
    'Together': 'Partnered', 
    'Married': 'Married', 
    'Divorced': 'Divorced', 
    'Widow': 'Widowed', 
    'Alone': 'Single', 
    'Absurd': 'Single',
    'YOLO': 'Single'
}
df['marital_status'] = df['marital_status'].map(marital_status_mapping)
```

```python
print(df['marital_status'].unique())
print(df['marital_status'].value_counts())
```

# Convert Date Formats & Fix Data Types

```python
df['dt_customer']
```

```python
# Convert Date Formats & Fix Data Types
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='mixed')
```

# Export the Cleaned Dataset

```python
df.to_csv("cleaned_customer_personality.csv", index=False)
print("Data cleaning complete! Cleaned dataset exported successfully.")
```
